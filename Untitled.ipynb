```python
# numpy and pandas
import numpy as np
import pandas as pd
from pandas import DataFrame, Series

# matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

titanic_data = pd.read_csv('./titanic_data.csv').head()
titanic_data.columns = map(str.lower, titanic_data.columns)
sns.factorplot('sex', kind='count', data=titanic_data)
```

```python
titanic_data.head()
```

```python
def ret_first_char(text):
    if type(text) == str and len(text) >= 1:
        return text[0]
    else:
        return np.nan

titanic_data['deck'] =  titanic_data['cabin'].apply(ret_first_char)
```

```python
titanic_data.head()
```

```python
cabin_df = DataFrame(titanic_data['deck'].dropna())
```
